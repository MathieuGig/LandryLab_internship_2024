```python
with open("../synth_genome/my_synthetic_genome.txt") as genome:
        genome_content = genome.read()
```

```python
# Removes first line puts a space
dna = re.sub(">.*", "", genome_content)
dna = re.sub("\n", "", dna)
```

```python
pattern = re.compile("N{600}")
matches = pattern.finditer(dna)

for i, match in enumerate(matches):
    print(f'span {i} : ', match.span())
    print(f'match {i} : ', match.group())
```
